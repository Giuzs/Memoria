```python
estados=["00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000", "00000000"]

operacao = input("Digite W para escrever, R para ler e qualquer tecla para parar: ")
while operacao == "R" or operacao == "W" or operacao == "L":
    if operacao == "W":
        endereco = input("Digite o endereço de 4 bits: ")
        dado = input("Digite o dado de 8 bits: ")

        decimal = 0
        expoente = 0
        for bit in reversed(endereco):
            if bit == "1":
                decimal = decimal + (2 ** expoente)
            expoente = expoente + 1
        estados.insert(decimal, dado)  

    elif operacao == "R":
        end= input("Digite o endereço: ")

        decimal = 0
        expoente = 0
        for bit in reversed(end):
            if bit == "1":
                decimal = decimal + (2 ** expoente)
            expoente = expoente + 1    
        print(estados[decimal])

    elif operacao == "L":
        print("Todos os estados:")
        for i in range (16):
            print(estados[i])

    operacao=input("Digite W para escrever, R para ler e qualquer tecla para parar: ")
```
